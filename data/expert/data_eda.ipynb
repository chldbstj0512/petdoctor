```python
import re

def clean_answer(text: str) -> str:
    if not isinstance(text, str):
        return text

    text = text.replace("\u200b", " ")

    patterns = [
        r"^안녕하세요\.?.*?입니다\.?",         
        r"하이닥.*?입니다\.?",               
        r"네이버\s*지식iN.*?입니다\.?",       
        r"감사합니다\.?",                      
        r"답변이\s*도움되시길\s*바라.*"      
    ]

    for p in patterns:
        text = re.sub(p, "", text, flags=re.IGNORECASE | re.DOTALL)

    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
df["answer_clean"] = df["answer"].apply(clean_answer)
```

```python
# Unnamed 컬럼 제거
df = df.drop(columns=["Unnamed: 0", "Unnamed: 0.1"], errors="ignore")

# CSV로 다시 저장
df.to_csv("/home/ys0660/happycat/data/data.csv", index=False)

print("✅ Unnamed columns removed and dataframe saved.")
```
